## Binning

<p>
    Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.

```python
df['number_of_q'] = df['number_of_q'].astype(float,copy=True)
```

```python
import matplotlib as plt
from matplotlib import pyplot

# draw histogram of attribute "number_of_questions" with bins = 3
plt.pyplot.hist(df['number_of_q'], bins = 3)

# set x, y labels and plot title
plt.pyplot.xlabel('number of questions?')
plt.pyplot.ylabel('count')
plt.pyplot.title('Number of questions bins')
```

```python
bins = np.linspace(min(df['number_of_q']), max(df['number_of_q']),4)
group_names = [ 'Low ', "Medium", "high"]
```

```python
df['number_of_q_binned'] = pd.cut(df['number_of_q'], bins, labels= group_names, include_lowest=True)
df[['number_of_q','number_of_q_binned']].head()
```

```python
df['number_of_q_binned'].value_counts()
```

```python
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df['number_of_q_binned'].value_counts())

# Set x/y labes and plot title
plt.pyplot.xlabel('number_of_q_binned')
plt.pyplot.ylabel('count')
plt.pyplot.title('number of questions bins')
```
